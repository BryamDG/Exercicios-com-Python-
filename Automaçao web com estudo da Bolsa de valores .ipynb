```python
import yfinance
import quantstats
import pandas
import matplotlib.pyplot as plt
import time

açao = input ("qual açao deseja analisar ? ")
destinatario = input("para quem deseja enviar o relatorio?")

dados = yfinance.Ticker(açao)
dados.history()
tabela = dados.history('3mo')
display(tabela)

fechamento = tabela.Close
#print (fechamento)


cotaçoes = yfinance.download(açao, interval="3mo")["Adj Close"]
retornos = cotaçoes.pct_change()

quantstats.extend_pandas()
retornos.plot_monthly_heatmap()

plt.show(block=True)
```

```python
graf=fechamento.plot ()
display (graf)

maxima = fechamento.max()
minima = fechamento.min()
atual = fechamento[-1]

print(f"este é o valor maximo da açao: {maxima:,.2f}, o minimo: {minima:,.2f} e o atual: {atual:,.2f} do fechamento dos ultimos 3 meses ")

```
